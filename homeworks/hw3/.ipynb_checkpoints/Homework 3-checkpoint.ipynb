## 2. Curve Parameterization

<a name='2-1'></a>
### 2.1. Equidistant (Uniform) Parameterization

```python
def uniform_parameterization(points):
    num_interval = len(points)
    return np.array([t / (num_interval-1) for t in range(num_interval)])
```

```python
t_EP = uniform_parameterization(points)
x_t_poly_interp_EP = interp_poly.interpolation_polynomial(t_EP, x_sample)
y_t_poly_interp_EP = interp_poly.interpolation_polynomial(t_EP, y_sample)
x_EP_poly=x_t_poly_interp_EP.interpolate(t_sample)
y_EP_poly=y_t_poly_interp_EP.interpolate(t_sample)

plt.plot(x_EP_poly, y_EP_poly, color='b')
plt.plot(x_sample, y_sample, 'o', color='r')
plt.show()
```

<a name='2-2'></a>
### 2.2. Chordal Parameterization

```python
def chordal_parameterization(points):
    intervals = np.array([0] + [np.linalg.norm(points[idx+1]-points[idx]) for idx in range(len(points)-1)])
    prefix_sum = np.cumsum(intervals)
    return prefix_sum / prefix_sum[-1]
```

```python
t_CP = chordal_parameterization(points)
x_t_poly_interp_CP = interp_poly.interpolation_polynomial(t_CP, x_sample)
y_t_poly_interp_CP = interp_poly.interpolation_polynomial(t_CP, y_sample)
x_CP_poly=x_t_poly_interp_CP.interpolate(t_sample)
y_CP_poly=y_t_poly_interp_CP.interpolate(t_sample)

plt.plot(x_CP_poly, y_CP_poly, color='b')
plt.plot(x_sample, y_sample, 'o', color='r')
plt.show()
```

<a name='2-3'></a>
### 2.3. Centripetal Parameterization

```python
def centripetal_parameterization(points):
    intervals = np.sqrt(np.array([0] + [np.linalg.norm(points[idx+1]-points[idx]) for idx in range(len(points)-1)]))
    prefix_sum = np.cumsum(intervals)
    return prefix_sum / prefix_sum[-1]
```

```python
t_CPSR = centripetal_parameterization(points)
x_t_poly_interp_CPSR = interp_poly.interpolation_polynomial(t_CPSR, x_sample)
y_t_poly_interp_CPSR = interp_poly.interpolation_polynomial(t_CPSR, y_sample)
x_CPSR_poly=x_t_poly_interp_CPSR.interpolate(t_sample)
y_CPSR_poly=y_t_poly_interp_CPSR.interpolate(t_sample)

plt.plot(x_CPSR_poly, y_CPSR_poly, color='y')
plt.plot(x_sample, y_sample, 'o', color='r')
plt.show()
```

<a name='2-4'></a>
### 2.3. Foley-Nielson Parameterization [[1]](#ref-1)

```python
def angle_vector(v1, v2):
    return np.arccos(np.inner(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

def foley_nielson_terms(angle, d1, d2):
    theta = min(np.pi-angle, np.pi/2)
    return 1.5 * theta * (d1/(d1+d2))
```

```python
def foley_nielson_parameterization(points):
    vectors = np.array([points[idx+1]-points[idx] for idx in range(len(points)-1)])
    distances = np.array([np.linalg.norm(v) for v in vectors])
    angles = np.array([angle_vector(vectors[idx], vectors[idx+1]) for idx in range(len(vectors)-1)])

    intervals = [0]
    for idx, distance in enumerate(distances):
        scalar = 1
        if not idx==0:
            scalar += foley_nielson_terms(angles[idx-1], distances[idx-1], distance)
        if not idx==len(distances)-1:
            scalar += foley_nielson_terms(angles[idx], distances[idx+1], distance)
        intervals.append(scalar*distance)
    
    prefix_sum = np.cumsum(np.array(intervals))
    return prefix_sum / prefix_sum[-1]
```

```python
t_FNP = foley_nielson_parameterization(points)
x_t_poly_interp_FNP = interp_poly.interpolation_polynomial(t_FNP, x_sample)
y_t_poly_interp_FNP = interp_poly.interpolation_polynomial(t_FNP, y_sample)
x_FNP_poly=x_t_poly_interp_FNP.interpolate(t_sample)
y_FNP_poly=y_t_poly_interp_FNP.interpolate(t_sample)

plt.plot(x_FNP_poly, y_FNP_poly, color='black')
plt.plot(x_sample, y_sample, 'o', color='r')
plt.show()
```
